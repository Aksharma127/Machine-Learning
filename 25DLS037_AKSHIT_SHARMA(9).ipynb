```python
from transformers import pipeline

# Load GPT-2 for text generation
generator = pipeline("text-generation", model="gpt2")

# Prompt and generation
prompt = "The future of AI is"
output = generator(prompt, max_new_tokens=30)

# Display output
print("Generated Text:\n", output[0]['generated_text'])
```

```python
# Low temperature (more deterministic)
output_low_temp = generator(prompt, max_new_tokens=30, temperature=0.2)
print("\nOutput with Temperature = 0.2:\n", output_low_temp[0]['generated_text'])

# High temperature (more random)
output_high_temp = generator(prompt, max_new_tokens=30, temperature=1.0)
print("\nOutput with Temperature = 1.0:\n", output_high_temp[0]['generated_text'])
```

```python
# Load BERT fill-mask pipeline
fill_mask = pipeline("fill-mask", model="bert-base-uncased")

# Sentence with masked token
masked_sentence = "The capital of India is [MASK]."

# Replace Hugging Face [MASK] token syntax
masked_sentence = masked_sentence.replace("[MASK]", fill_mask.tokenizer.mask_token)

# Get top 2 predictions
results = fill_mask(masked_sentence, top_k=2)

# Print top 2 results
print("\nTop 2 Predictions for Masked Token:")
for result in results:
    print(f"{result['sequence']}  (Score: {result['score']:.4f})")
```
