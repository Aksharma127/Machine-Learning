```python
critical_columns = ['DBA', 'BUILDING', 'STREET', 'ZIPCODE', 'PHONE', 'CUISINE DESCRIPTION', 'ACTION', 'VIOLATION CODE', 'VIOLATION DESCRIPTION', 'SCORE', 'GRADE', 'GRADE DATE', 'INSPECTION TYPE', 'Latitude', 'Longitude', 'Community Board', 'Council District', 'Census Tract', 'BIN', 'BBL', 'NTA']
df_clean.dropna(subset=critical_columns, inplace=True)
print(f"Number of remaining rows after dropping missing values: {len(df_clean)}")
```

```python
df_clean['inspection_year'] = df_clean['INSPECTION DATE'].dt.year
df_clean['violation_category'] = df_clean['CRITICAL FLAG'].apply(lambda x: 'critical violation' if x == 'critical' else 'other violation')
display(df_clean.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

def run_full_analysis(file_path):
    """
    Executes the entire data cleaning and analysis pipeline
    from loading the raw data to generating final plots.
    """

    # Suppress warnings for a cleaner output
    warnings.filterwarnings('ignore')

    # --- PHASE 1: LOAD DATA ---
    print("--- Phase 1: Loading Data ---")
    try:
        # df_raw will be loaded from the file_path provided
        df_raw = pd.read_csv(file_path)
        print(f"Successfully loaded {file_path}")
        print(f"Raw data shape: {df_raw.shape}")
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        print("Please find the file in the Colab file panel, right-click, select 'Copy path', and paste it into the 'file_to_analyze' variable.")
        return
    except Exception as e:
        print(f"An error occurred loading the file: {e}")
        return

    # --- PHASE 3: CLEANING & FEATURE ENGINEERING ---
    print("\n--- Phase 3: Cleaning Data ---")

    # 1. Create a copy
    df_clean = df_raw.copy()

    # 2. Fix Data Types
    df_clean['SCORE'] = pd.to_numeric(df_clean['SCORE'], errors='coerce')
    df_clean['INSPECTION DATE'] = pd.to_datetime(df_clean['INSPECTION DATE'], errors='coerce')
    print("Fixed data types for 'SCORE' and 'INSPECTION DATE'.")

    # 3. Handle Missing Values
    initial_rows = df_clean.shape[0]
    # Drop rows where core columns are missing
    df_clean.dropna(subset=['SCORE', 'INSPECTION DATE', 'BORO', 'DBA'], inplace=True)
    print(f"Dropped {initial_rows - df_clean.shape[0]} rows with missing critical data.")

    # 4. Handle Garbage Values
    initial_rows = df_clean.shape[0]
    # Remove '1900' dates
    df_clean = df_clean[df_clean['INSPECTION DATE'].dt.year >= 2000]
    # Remove negative scores
    df_clean = df_clean[df_clean['SCORE'] >= 0]
    print(f"Removed {initial_rows - df_clean.shape[0]} rows with garbage dates or scores.")

    # 5. Standardize Text
    text_cols = ['DBA', 'CUISINE DESCRIPTION', 'BORO']
    for col in text_cols:
        df_clean[col] = df_clean[col].str.lower().str.strip()
    print("Standardized text for 'DBA', 'CUISINE DESCRIPTION', and 'BORO'.")

    # 6. Feature Engineering
    df_clean['inspection_year'] = df_clean['INSPECTION DATE'].dt.year

    # Create violation_category based on 'VIOLATION DESCRIPTION'
    df_clean['violation_category'] = 'Other/Minor'
    df_clean.loc[df_clean['VIOLATION DESCRIPTION'].str.contains('mice|rodent', case=False, na=False), 'violation_category'] = 'Pest Control Issues'
    df_clean.loc[df_clean['VIOLATION DESCRIPTION'].str.contains('food temperature|temperature abuse', case=False, na=False), 'violation_category'] = 'Food Temp Control'
    df_clean.loc[df_clean['VIOLATION DESCRIPTION'].str.contains('personal hygiene|hand washing|gloves', case=False, na=False), 'violation_category'] = 'Staff Hygiene'
    print("Engineered 'violation_category' from 'VIOLATION DESCRIPTION'.")

    print(f"Cleaning complete. Final data shape: {df_clean.shape}")

    # --- PHASE 4: "AFTER" ANALYSIS & VISUALIZATION ---
    print("\n--- Phase 4: Generating 'After' Analysis & Plots ---")

    sns.set_theme(style="whitegrid")

    # 1. Calculate and print clean average score
    avg_score_clean = df_clean['SCORE'].mean()
    print(f"\nAverage Score (Cleaned): {avg_score_clean:.2f}")

    # 2. Calculate and print average score by borough
    print("\n--- Average Score by Borough (Cleaned) ---")
    avg_score_by_borough = df_clean.groupby('BORO')['SCORE'].mean().sort_values()
    print(avg_score_by_borough)

    # 3. Plot Top 5 Violations (using the new category)
    plt.figure(figsize=(10, 6))
    top_violations_cat = df_clean['violation_category'].value_counts().head(5)
    sns.barplot(x=top_violations_cat.values, y=top_violations_cat.index)
    plt.title('Top 5 Violation Categories (Cleaned Data)')
    plt.xlabel('Number of Violations')
    plt.tight_layout()
    plt.savefig('clean_top_5_violations.png')
    plt.show()
    print("\nGenerated 'Top 5 Violations' plot.")

    # 4. Plot Scores Over Time
    plt.figure(figsize=(12, 6))
    df_clean.groupby('inspection_year')['SCORE'].mean().plot(kind='line', marker='o')
    plt.title('Average Inspection Score Over Time (Cleaned Data)')
    plt.ylabel('Average Score')
    plt.xlabel('Year')
    plt.grid(True)
    plt.savefig('clean_scores_over_time.png')
    plt.show()
    print("Generated 'Scores Over Time' plot.")

    # 5. Plot Score Distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(df_clean['SCORE'], bins=40, kde=True)
    plt.title('Distribution of Restaurant Scores (Cleaned Data)')
    plt.xlabel('Score')
    plt.axvline(avg_score_clean, color='red', linestyle='--', label=f'Mean Score: {avg_score_clean:.2f}')
    plt.legend()
    plt.savefig('clean_score_distribution.png')
    plt.show()
    print("Generated 'Score Distribution' plot.")

    # 6. Plot Score by Borough
    plt.figure(figsize=(12, 7))
    sns.boxplot(data=df_clean, x='BORO', y='SCORE', order=avg_score_by_borough.index)
    plt.title('Score Distribution by Borough (Cleaned Data)')
    plt.savefig('clean_score_by_borough.png')
    plt.show()
    print("Generated 'Score by Borough' plot.")

    print("\n--- Full Analysis Complete ---")

# --- !! IMPORTANT !! ---
# --- Replace this path with the one you copied from your Colab file panel ---
file_to_analyze = "/content/sample_data/DOHMH_New_York_City_Restaurant_Inspection_Results_20251027.csv"
# -------------------------

run_full_analysis(file_to_analyze)
```
