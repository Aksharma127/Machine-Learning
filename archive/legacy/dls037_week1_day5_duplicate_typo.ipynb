Each probability is between 0 and 1, and all of them add up to 1. It's often used in classification tasks.**Formula:**$$\text{softmax}(z_i) = \frac{e{z_i}}{\sum_j e{z_j}}$$

```python
import numpy as npdef simple_softmax(x_values):    exponentials = np.exp(x_values - np.max(x_values))    return exponentials / np.sum(exponentials)    # Testinput_numbers = np.array([2.0, 2.8, 3.1])probabilities = simple_softmax(input_numbers)print(f"Input: {input_numbers}")print(f"Softmax Output: {probabilities}")print(f"Sum: {np.sum(probabilities)}")
```

---## **P-3: PyTorch and NumPy Basics [5 Marks]****The Task:**1. Make a 5x5 matrix of random numbers (0 to 1) with PyTorch.2. Find its mean, standard deviation, and sum using PyTorch.3. Convert it to a NumPy array.4. Repeat the same calculations using NumPy.

```python
import torchimport numpy as npprint("--- Using PyTorch ---")# PyTorch: Create a 5x5 random matrixpt_matrix = torch.rand(5, 5)print("PyTorch Matrix:\n", pt_matrix)# PyTorch: Calculate statspt_mean = torch.mean(pt_matrix)pt_std = torch.std(pt_matrix)pt_sum = torch.sum(pt_matrix)print(f"\nPyTorch Mean: {pt_mean.item()}")print(f"PyTorch Std Dev: {pt_std.item()}")print(f"PyTorch Sum: {pt_sum.item()}")print("\n--- Using NumPy ---")# Convert to NumPy arraynp_array = pt_matrix.numpy()print("NumPy Array (from PyTorch):\n", np_array)# NumPy: Calculate statsnp_mean = np.mean(np_array)np_std = np.std(np_array)np_sum = np.sum(np_array)print(f"\nNumPy Mean: {np_mean}")print(f"NumPy Std Dev: {np_std}")print(f"NumPy Sum: {np_sum}")
```
