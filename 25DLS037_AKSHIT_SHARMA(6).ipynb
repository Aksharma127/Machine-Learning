## Task 1 – TensorFlow Implementation (5 Marks)


```python
import tensorflow as tf

# Input (1, 3)
X = tf.constant([[0.5, 1.5, -2.0]], dtype=tf.float32)

# Manually initialized weights (3, 1) and bias (scalar)
W = tf.Variable([[0.2], [0.4], [-0.5]], dtype=tf.float32)
b = tf.Variable([0.1], dtype=tf.float32)

z = tf.matmul(X, W) + b

y_pred = tf.math.sigmoid(z)
y_true = tf.constant([[1.0]], dtype=tf.float32)
epsilon = 1e-7
loss = - (y_true * tf.math.log(y_pred + epsilon) + (1 - y_true) * tf.math.log(1 - y_pred + epsilon))

print("Predicted:", y_pred.numpy())
print("Loss:", loss.numpy())
```

## Task 2 – PyTorch Implementation (5 Marks)


```python
import torch

# Input (1, 3)
X = torch.tensor([[0.5, 1.5, -2.0]], dtype=torch.float32)

# Manually initialized weights and bias (requires_grad=True)
W = torch.tensor([[0.2], [0.4], [-0.5]], dtype=torch.float32, requires_grad=True)
b = torch.tensor([0.1], dtype=torch.float32, requires_grad=True)

# Compute z = XW + b
z = torch.mm(X, W) + b

# Apply sigmoid activation
y_pred = torch.sigmoid(z)
y_true = torch.tensor([[1.0]], dtype=torch.float32)
epsilon = 1e-7
loss = - (y_true * torch.log(y_pred + epsilon) + (1 - y_true) * torch.log(1 - y_pred + epsilon))

# Backward pass
loss.backward()

print("Predicted:", y_pred.detach().numpy())
print("Loss:", loss.item())
print("Weight Gradients:", W.grad.numpy())
print("Bias Gradients:", b.grad.numpy())
```
