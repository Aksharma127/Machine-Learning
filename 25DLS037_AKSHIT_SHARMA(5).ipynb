### 2. Eigenvalues and Eigenvectors
Define matrix D and compute its eigenvalues and eigenvectors using NumPy. Plot the eigenvectors using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

D = np.array([[4, 2], [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(D)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Plotting eigenvectors
# Define the origin for each vector
origin_x = [0, 0]
origin_y = [0, 0]

# Plotting eigenvectors
# Pass the origin points as arrays
plt.quiver(origin_x, origin_y, eigenvectors[:, 0], eigenvectors[:, 1], color=['r', 'b'], scale=3, angles='xy', scale_units='xy')
plt.grid()
plt.title("Eigenvectors of Matrix D")
plt.axis("equal")
plt.show()
```

## P-2: Symbolic Differentiation and Gradient Step [3 Marks]

### 1. Symbolic Differentiation
Use `sympy` to define the function $f(x) = 3x^3 - 5x^2 + 2x - 1$, find its derivative, and evaluate it at $x=2$.

```python
import sympy as sp

x = sp.symbols('x')
f = 3*x**3 - 5*x**2 + 2*x - 1
f_prime = sp.diff(f, x)

print("Function:", f)
print("Derivative:", f_prime)
print("Derivative at x=2:", f_prime.subs(x, 2))
```

### 2. Gradient and Descent
Define $f(x, y) = x^2 + y^2$, compute its gradient, and perform one gradient descent step from $(x=3, y=4)$ using learning rate = 0.1

```python
x, y = sp.symbols('x y')
f_xy = x**2 + y**2

# Gradient
grad_f = [sp.diff(f_xy, var) for var in (x, y)]
print("Gradient:", grad_f)

# Gradient descent step
learning_rate = 0.1
point = np.array([3.0, 4.0])
grad_evaluated = np.array([grad.evalf(subs={x: point[0], y: point[1]}) for grad in grad_f], dtype=float)
new_point = point - learning_rate * grad_evaluated

print("Original point:", point)
print("Gradient at point:", grad_evaluated)
print("New point after one step:", new_point)
```

## P-3: Probability with Dice and Coin Simulation [4 Marks]

### 1. Dice Simulation
Simulate rolling a fair die 1000 times, plot histogram, and print empirical mean and variance.

```python
rolls = np.random.randint(1, 7, size=1000)

# Histogram
plt.hist(rolls, bins=np.arange(1, 8)-0.5, edgecolor='black')
plt.title("Histogram of 1000 Dice Rolls")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.xticks(range(1, 7))
plt.show()

# Empirical stats
mean_roll = np.mean(rolls)
var_roll = np.var(rolls)
print("Empirical Mean:", mean_roll)
print("Empirical Variance:", var_roll)
```

### 2. Biased Coin Toss
Simulate a biased coin toss (P(heads)=0.7) 500 times, count outcomes, and plot bar chart.

```python
coin_tosses = np.random.choice(['Heads', 'Tails'], size=500, p=[0.7, 0.3])
heads = np.count_nonzero(coin_tosses == 'Heads')
tails = np.count_nonzero(coin_tosses == 'Tails')

# Bar chart
plt.bar(['Heads', 'Tails'], [heads, tails], color=['gold', 'silver'])
plt.title("Biased Coin Toss (P(Heads)=0.7)")
plt.ylabel("Count")
plt.show()

print("Heads:", heads)
print("Tails:", tails)
```
